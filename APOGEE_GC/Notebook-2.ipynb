```python
import astropy.io.fits as fits
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
```

```python

filename = 'GC_parameters_VAC-v1_1.fits'
with fits.open(filename) as hdul:
    # Print the information about the FITS file
    hdul.info()

    # Access the first data table (usually found in the first extension)
    data = hdul[1].data

    # Convert the data to a structured array


    # If you want to store all columns in a dictionary
    columns_data = {col.name: data[col.name] for col in data.columns}

# second__filename = 'GC_members_VAC-v1_1.fits'
# with fits.open(second_filename) as hdul:
#     # Print the information about the FITS file
#     hdul.info()

#     # Access the first data table (usually found in the first extension)
#     data = hdul[1].data

#     # Convert the data to a structured array
```

```python
print(columns_data.keys())
```

```python
# getting median metallicity and standard deviation of metallicity for a given cluster from second file
def get_data(cname):
    idx = np.where(columns_data['GC_name'] == cname)
    if len(idx) == 0:
        print(f"Cluster {cname} not found.")
        return
    Fe_H_med, Fe_H_sig = columns_data['Fe_H_med'][idx], columns_data['Fe_H_sig'][idx]
    print(f"Median metallicity (Fe/H) for {cname}: {Fe_H_med}")
    print(f"Standard deviation of metallicity (Fe/H) for {cname}: {Fe_H_sig}")
```

```python
get_data('NGC0104')
```
